```python
model_id = "unsloth/llama-3-8b-Instruct-bnb-4bit"
pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={
        "torch_dtype": torch.float16,
        "quantization_config": {"load_in_4bit": True},
        "low_cpu_mem_usage": True,
    },
)
```

```python
def generate_prompt(title):
    messages = [
        {"role": "system", "content": "You are a youtube video classifier! You only reply with one word, the generalized category of all the video titles"},
        {"role": "user", "content": f"Find the category that best fits the following video titles: {title}"},
    ]
    prompt = pipeline.tokenizer.apply_chat_template(
        messages, 
        tokenize=False, 
        add_generation_prompt=True
    )
    terminators = [
        pipeline.tokenizer.eos_token_id,
        pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
    ]
    outputs = pipeline(
        prompt,
        max_new_tokens=256,
        eos_token_id=terminators,
        do_sample=True,
        temperature=0.6,
        top_p=0.9,
    )
    print(outputs[0]["generated_text"][len(prompt):])
    return outputs[0]["generated_text"][len(prompt):]
```

```python
def standardize_to_plural(df):
    unique_categories = df['category'].unique()
    
    for category in unique_categories:
        singular_form = category.rstrip('s')
        plural_form = singular_form + 's'
        
        if plural_form in unique_categories:
            df['category'] = df['category'].replace(singular_form, plural_form)
    
    return df
```

```python
import re
def convert_to_number(text):
    if isinstance(text, str):
        text = re.sub(r'[^\d\.KkMm]', '', text)
        text = text.replace(',', '')

        if 'k' in text.lower():
            return float(text.lower().replace('k', '')) * 1_000
        elif 'm' in text.lower():
            return float(text.lower().replace('m', '')) * 1_000_000
        return float(text)
    return text  
```

```python
from datetime import datetime

def extract_date(text):
    text = text.replace('Joined ', '')  
    
    date_formats = [
        '%b %d, %Y', '%d %b %Y', '%Y-%m-%d', 
        '%d %b %Y', '%d %B %Y', '%B %d, %Y',
        '%d %B %Y', '%d %b %Y', '%d %m %Y'
    ]
    
    for date_format in date_formats:
        try:
            return datetime.strptime(text, date_format)
        except ValueError:
            continue
    
    
    raise ValueError(f"Date format not recognized: {text}")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('final.csv')
```

```python
scaler = StandardScaler()
numerical_features = ['subscribers', 'videos', 'views']
data[numerical_features] = scaler.fit_transform(data[numerical_features])
```

```python
encoder = OneHotEncoder()
categories_encoded = encoder.fit_transform(data[['category']]).toarray()
```

```python
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
description_tfidf = tfidf_vectorizer.fit_transform(data['description']).toarray()
titles_tfidf = tfidf_vectorizer.fit_transform(data['titles']).toarray()
```

```python
combined_features = np.hstack([data[numerical_features], categories_encoded, description_tfidf, titles_tfidf])
```

```python
# Cosine similarity between all channels
similarity_matrix = cosine_similarity(combined_features)
```

```python
def find_similar_channels(channel_index, similarity_matrix, df, top_n=5):
    # Get the row corresponding to the input channel
    channel_similarities = similarity_matrix[channel_index]
    
    # Get the indices of the top N similar channels
    similar_indices = np.argsort(-channel_similarities)[1:top_n+1]
    
    # Return the similar channels
    return df.iloc[similar_indices]
```

# Example usage:

```python
input_channel_index = 5  #Index of the input channel
similar_channels = find_similar_channels(input_channel_index, similarity_matrix, data, top_n=10)
print(similar_channels[['url', 'subscribers', 'views', 'videos', 'category']])
```
